## Environemnt
This is a real hell, bugs with various versions and soo on. I'll try to explain how I got it to work at this moment in time....
1. Install Miniconda or Anaconda
2. Add source channel: conda config --add channels conda-forge  
3. Add source channel: conda config --add channels Jupyter 
4. Create and activate an new environment, I used 3.10
   - conda create --name=machinelearning python=3.10
5. conda install ipykernel jupyterlab notebook voila 
6. conda install openpyxl sqlalchemy scikit-learn missingno ipywidgets
   - **Missingno is not updated to use the new matplotlib, downgrade to older matplotlib**
   - conda install matplotlib=3.6.3 
7. conda install ydata-profiling
   - **ProfileReport was not comaptible with latest numpy, downgrade**
   - conda install numpy<1.25  
   - Alternative: pip install --force-reinstall -v "numpy<1.25"
     - Men seaborn fungerar inte
       - conda install numpy>1.25 - men såklart fungerar inte ProfileReport... 
8. conda install pandera
   - **Incompatible with typeing-extension 4.8.0 **
      - try typing-extensions==4.5.0 -> pandera works BUT ydata_profiling fail
      - try typing-extensions==4.6.0 -> pandera fail AND ydata_profiling works 
      - Conclusion: use latest typing-extensions and downgrade to 4.5.0 if you need pandera and schema verification
9. conda install seaborn
   - **Incompatible with numpy 1.25**
      - try numpy<1.20 -> seaborn works AND ProfileReport works
