```python
sample = services["Namn"].drop_duplicates() #.sample(10)
dataview = df[df['Namn'].isin(sample)]
dataview = dataview.drop(['Ägare', 'Ansvarig','Förvaltningsobjekt', 'Supportgrupp (3rd)', 'Supportgrupp (2nd)'], axis=1)
dataview = dataview.rename(columns={"Namn": "Service", "Grupp": "Team", "E-post": "E-post"})
dataview
```

```python
serviceteams = dataview.drop(['E-post'], axis=1)
serviceteams = serviceteams.drop_duplicates()
for i, row in serviceteams.iterrows():
    antal = serviceteams[serviceteams["Team"] ==serviceteams.at[i,"Team"]]["Team"].count()
    serviceteams.at[i,'Amount'] = 1/antal
serviceteams
```

```python
import networkx as nx
import numpy as np
from pyvis.network import Network

team_members = pd.DataFrame(
    {
        "source": serviceteams["Service"].tolist(),
        "target": serviceteams["Team"].tolist(),
        "weight": serviceteams["Amount"].tolist()
    }
)
G_team_members = nx.from_pandas_edgelist(team_members, edge_attr=True)
net = Network(notebook=True)
for source in team_members['source'].unique():
    net.add_node(source, label=source)
for target in team_members['target'].unique():
    net.add_node(target, label=target, color='red')
for index, row in team_members.iterrows():
    net.add_edge(row['source'], row['target'], value=row['weight']/3, color='gray')
net.show("service_teams.html")
```

```python
with open("service_teams.html", "r") as file:
    html_data = file.read()

start = html_data.find("nodes =")
end = html_data.find("nodeColors = {};")
javascript_data = html_data[start:end].strip()
javascript_data = javascript_data.replace("nodes = ", "service_nodes = ")
javascript_data = javascript_data.replace("edges = ", "service_edges = ")
file = open("data/service_teams.js", "wt") #(r)ead, (a)ppend, (w)rite, create(x) (t)ext/(b)inary
file.write(javascript_data)
file.close()

import os
os.remove("service_teams.html")
```
