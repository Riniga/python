## Merge multiple of events into one dataframe

```python
events = list()
events = list()
events.append({"name": "Skåneserien 3 2021 Skurup", "url": "https://svenskjudo.smoothcomp.com/sv/event/3880/results"})
events.append({"name": "Skåneserien 4 2021 Staffanstorp", "url": "https://svenskjudo.smoothcomp.com/sv/event/5516/results"})
events.append({"name": "Skåneserien 1 2022 Skurup", "url": "https://svenskjudo.smoothcomp.com/sv/event/6424/results"})
events.append({"name": "Skåneserien 2 2022 Skurup", "url": "https://svenskjudo.smoothcomp.com/sv/event/7314/results"})
events.append({"name": "Skåneserien 3 2022 Knislinge", "url": "https://svenskjudo.smoothcomp.com/sv/event/8558/results"})
events.append({"name": "Skåneserien 4 2022 Staffanstorp", "url": "https://svenskjudo.smoothcomp.com/sv/event/6838/results"})
events.append({"name": "Skåneserien 2 2023 Knislinge", "url": "https://svenskjudo.smoothcomp.com/sv/event/9302/results"})
events.append({"name": "Skåneserien 3 2023 Skurup", "url": "https://svenskjudo.smoothcomp.com/sv/event/10180/results"})
events.append({"name": "Skåneserien 4 2023 Staffanstorp", "url": "https://svenskjudo.smoothcomp.com/sv/event/12871/results"})
events.append({"name": "Skåneserien 1 2024 Staffanstorp", "url": "https://svenskjudo.smoothcomp.com/sv/event/14644/results"})
events.append({"name": "Skåneserien 2 2024 Knislinge",    "url": "https://svenskjudo.smoothcomp.com/en/event/12968/results"}) 

frames = list()
for event in events:
    frames.append(pd.read_csv(event["name"] + '_result.csv'))
df = pd.concat(frames)

pd.set_option('display.max_rows', df.shape[0]+1)

df
```

```python
df.to_csv('all_results.csv', index=False, encoding='utf-8-sig')# cp1252  iso-8859-1 utf-8 
```
