```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

## Merge multiple of events into one dataframe

```python
events = list()
events.append({"name": "Skåneserien 3 2021 Skurup", "url": "https://svenskjudo.smoothcomp.com/sv/event/3880/schedule/matchlist"})
events.append({"name": "Skåneserien 4 2021 Staffanstorp", "url": "https://svenskjudo.smoothcomp.com/sv/event/5516/schedule/matchlist"})
events.append({"name": "Skåneserien 1 2022 Skurup", "url": "https://svenskjudo.smoothcomp.com/sv/event/6424/schedule/matchlist"})
events.append({"name": "Skåneserien 2 2022 Skurup", "url": "https://svenskjudo.smoothcomp.com/sv/event/7314/schedule/matchlist"})
events.append({"name": "Skåneserien 3 2022 Knislinge", "url": "https://svenskjudo.smoothcomp.com/sv/event/8558/schedule/matchlist"})
events.append({"name": "Skåneserien 4 2022 Staffanstorp", "url": "https://svenskjudo.smoothcomp.com/sv/event/6838/schedule/matchlist"})
events.append({"name": "Skåneserien 2 2023 Knislinge", "url": "https://svenskjudo.smoothcomp.com/sv/event/9302/schedule/matchlist"})
events.append({"name": "Skåneserien 3 2023 Skurup", "url": "https://svenskjudo.smoothcomp.com/sv/event/10180/schedule/matchlist"})
events.append({"name": "Skåneserien 4 2023 Staffanstorp", "url": "https://svenskjudo.smoothcomp.com/sv/event/12871/schedule/matchlist"})
events.append({"name": "Skåneserien 1 2024 Staffanstorp", "url": "https://svenskjudo.smoothcomp.com/sv/event/14644/schedule/matchlist"})
events.append({"name": "Skåneserien 2 2024 Knislinge",    "url": "https://svenskjudo.smoothcomp.com/en/event/12968/schedule/matchlist"})

frames = list()
for event in events:
    frames.append(pd.read_csv(event["name"] + '.csv'))
df = pd.concat(frames)
df
```

```python
df.to_csv('all.csv', index=False, encoding='utf-8-sig')# cp1252  iso-8859-1 utf-8 
```

## Create statistics

```python
df_white = df[["White Name" ,"White Club"]].rename(columns={"White Name": "Name", "White Club": "Club"})
df_blue = df[["Blue Name" ,"Blue Club"]].rename(columns={"Blue Name": "Name", "Blue Club": "Club"})
df_players = pd.concat([df_white, df_blue])
df_players.drop_duplicates(inplace = True)
df_players
```

```python
Wins = df.groupby(["Competition", "Winner Club", "Winner Name" ]).count().reset_index()[["Competition", "Winner Club", "Winner Name" ,"Winner Color"]].rename(columns={"Winner Club": "Club", "Winner Name":"Name",  "Winner Color": "Wins"})
Losses = df.groupby(["Competition", "Looser Club", "Looser Name" ]).count().reset_index()[["Competition", "Looser Club", "Looser Name" ,"Winner Color"]].rename(columns={"Looser Club": "Club", "Looser Name":"Name",  "Winner Color": "Losses"})

df_result = pd.merge(Wins, Losses, how='outer')
df_result = df_result.fillna(0)

df_result["Win Ratio"] = df_result["Wins"] /( df_result["Wins"]+ df_result["Losses"])
df_result.sort_values(["Win Ratio"], ascending=False)

```

```python
ClubResult = df_result.groupby(["Club"]).sum().reset_index()[["Club", "Wins" ,"Losses"]]
ClubResult["Win Ratio"] = ClubResult["Wins"] /( ClubResult["Wins"]+ ClubResult["Losses"])
ClubResult['Win Ratio'] = ClubResult['Win Ratio'].map('{:.0%}'.format)
ClubResult.sort_values(["Win Ratio"], ascending=False)
```
